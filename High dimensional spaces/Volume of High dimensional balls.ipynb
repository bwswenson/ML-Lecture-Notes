This code plots the volume of a high dimensional ball vs dimension. The volume of the ball is given by
$$
V_d(r) = \frac{\pi^{d/2} r^d}{\Gamma(d/2 + 1)}. 
$$
First a standard plot is given then a log plot

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
from scipy.special import gamma

def V(d, r):
    return pi**(d/2)*r**d/gamma(d/2+1)

N = 25
Z = [z for z in range(N)]
y1 = [V(z, 1) for z in Z]
y9 = [V(z, .9) for z in Z]
y11 = [V(z, 1.1) for z in Z]
plt.plot(Z, y1, Z, y9, Z, y11)
plt.legend(('r=1', 'r=.9', 'r=1.1'))
plt.show()
```

Next, we see this in log land, to see how fast it really decays. 

```python
N = 300
Z = [z for z in range(N)]
y1 = [V(z, 1) for z in Z]
y9 = [V(z, .9) for z in Z]
y11 = [V(z, 1.1) for z in Z]
plt.plot(Z,np.log(y1), Z,np.log(y9), Z,np.log(y11))
plt.legend(('r=1', 'r=.9', 'r=1.1'))
plt.show()
```

The counterintuitive property (imo) is that the volume of the ball eventually goes to zero even if the radius is arbitrarily larger than 1. 

If the radius of the ball is bigger, it contains more volume and takes longer to kill in the dimension. Remember that the plot below is logarathmic, so a value of y means the volume is something raised to the y. But the factorial always eventually wins out.

```python
N = 300
r = 6
Z = [z for z in range(N)]
y1 = [V(z, r) for z in Z]
# y9 = [V(z, 2) for z in Z]
# y11 = [V(z, 5) for z in Z]
plt.plot(Z,np.log(y1))
plt.legend(('r=1', 'r=2', 'r=5'))
plt.show()
```
