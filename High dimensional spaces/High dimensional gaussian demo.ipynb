This code demos the fact that a $d$-dimensional gaussian has most of its mass concentrated on the surface of a sphere of radius $\sqrt{d-1}$. First, plot the pdf of a high dimensional Gaussian, radially. 

The change of variables formula tells us that if $Y = g(X)$, then 
$$
f_Y(y) = f_X(g^{-1}(y)) |\det D g^{-1}(y)|.
$$ 
In particular, if $Y=\sqrt{X}$ we get
$$
f_Y(y) = 2yf_X(y^2).
$$

Let $Z = (Z_i)_{i=1}^d$ be a unit normal RV. Let $r = \|Z\|$. We're interested in computing the density of $r$. Note that $r^2$ is chi2 distributed. That is, $r =\sqrt{W}$ for $W$ chi2 distributed. Letting $f_W(\cdot, d)$, denote the chi2 distribution with $d$ degrees of freedom and applying the above, we see that
$$
f_r(r) = 2rf_W(r^2, d).
$$

Hence, we may plot the closed form density of $r$ below using the built in density for chi2 (thus avoiding getting bogged down in numerical issues). 

Play around with $d$ below to plot the radial density of the Gaussian. 

```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = 2000
f = chi2.pdf  # takes as second argument degrees of freedom

x = np.linspace(0,50, 1000)
plt.plot(x, f(x**2, d))
```

To see this from a sampling perspective, sample from a unit gaussian and plot the histogram of $\|Z\|$. 

```python
N = 1000 # num samples
Z = np.random.normal(size=(d,N))
r = np.linalg.norm(Z, axis=0)
plt.hist(r)
```
